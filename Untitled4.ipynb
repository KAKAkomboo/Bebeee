```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay as Matdis
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("Pokemon.csv") #Завантажуємо файл
```

```python
df
```

```python
df = df[["Type 1", "HP", "Attack", "Defense", "Speed", "Total"]] #Вибираємо потрібні стовпці
```

```python
df
```

```python
y = df["Type 1"].values #Цільова змінна (Тип покемона)
```

```python
y
```

```python
x = df[["HP", "Attack", "Defense", "Speed", "Total"]].values #Характеристики покемона
```

```python
x
```

```python
df["Type 1"].unique() #Переглядаємо типи покемонів
```

```python
en = sklearn.preprocessing.OneHotEncoder()
```

```python
y = en.fit_transform(y.reshape(-1, 1)) # Тип покемона перетворюємо на числа. Кожен тип стає набором із 0 та 1
```

```python
y = y.toarray()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42) #Розділяємо дані на тренувальні та тестові
```

```python
print(len(x_train), len(x_test), len(y_train), len(y_test))
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train) # Наприклад, якщо одна характеристика, наприклад, Total має великі значення, а інша, наприклад, Speed - малі, це може впливати на модель, а нормалізація це виправляє
```

```python
x_train
```

```python
classifier = DecisionTreeClassifier(random_state = 42, max_depth = 5, max_leaf_nodes = 12) # Створюєм та навчаєм модель дерева рішень. max_depth: Максимальна глибина дерева, max_leaf_nodes: Максимальна кількість вузлів
```

```python
classifier
```

```python
classifier.fit(x_train, y_train)
```

```python
y_test_pred = classifier.predict(x_test) # Модель намагається вгадати тип покемонів, використовуючи тестові дані
```

```python
print(classification_report(y_test, y_test_pred)) # Виводиться звіт, який показує, наскільки добре модель вгадала типи.
```

```python
y_test
```

```python
y_test_by_axsi_1 = np.argmax(y_test, axis = 1)
```

```python
y_test_pred_by_axsi_1 = np.argmax(y_test_pred, axis = 1)
```

```python
confusion_array = confusion_matrix(y_test_by_axsi_1, y_test_pred_by_axsi_1)
matdis = Matdis(confusion_array, display_labels = en.categories_[0]) 
matdis.plot() # Створюємо матрицю плутанини, щоб подивитися, які типи покемонів найчастіше плутаються
```

```python
accuracy_score(y_test, y_test_pred)
```

```python
plt.figure(figsize = (30, 10))
plot_tree(classifier, feature_names = df.columns, filled = True, fontsize = 6, impurity = False)
plt.show() # Малюємо дерево рішень, щоб побачитии, як модель приймає рішення
```

<h1>RandomForestClassifier</h1>

```python
rfc = RandomForestClassifier(criterion = "entropy", random_state = 42) # Використовуючи RandomForestClassifier ми може дізнатись більш точний результат
```

```python
rfc.fit(x_train, y_train)
```

```python
y_test_pred = rfc.predict(x_test)
print(classification_report(y_test, y_test_pred)) # Робимо передбачення і перевіряємо наскільки добре модель працює
```
